```python
import pandas as pd
```

```python
emp_sal = pd.read_csv("C:/Users/javed/Downloads/Data-cleaning-for-beginners-using-pandas.csv")
```

```python
emp_sal["Age"].fillna(emp_sal["Age"].mean(),inplace = True)
```

```python
emp_sal["Age"] = emp_sal["Age"].astype(int)
```

```python
def sal_conversion(Salary):
    Salary = Salary.replace("$", "")
    sal_1 = int(Salary.split("-")[0][:-1]) * 1000
    sal_2 = int(Salary.split("-")[1][:-1]) * 1000
    return f"${sal_1}-{sal_2}"

emp_sal["Salary"] = emp_sal["Salary"].apply(sal_conversion)
```

```python
emp_sal["Rating"].fillna(emp_sal["Rating"].median(),inplace = True)
```

```python
emp_sal["Rating"]=emp_sal["Rating"].abs()
```

```python
def Location(arg):
    country = arg.split(",")[0]
    return country
emp_sal["Location"]=emp_sal["Location"].apply(Location)
```

```python
emp_sal["Established"] = emp_sal["Established"].replace(-1,emp_sal["Established"].median())
```

```python
emp_sal["Easy Apply"] = emp_sal["Easy Apply"].replace("-1","FALSE")
```
