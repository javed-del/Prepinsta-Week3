```python
import pandas as pd
```

```python
chipotle_df = pd.read_csv("chipotle.tsv",sep='\t')
```

```python
chipotle_df['choice_description'] = chipotle_df['choice_description'].str.replace("[","")
```

```python
chipotle_df['choice_description'] = chipotle_df['choice_description'].str.replace("]","")
```

```python
chipotle_df = chipotle_df.dropna(subset="choice_description")
```

```python
chipotle_df
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(chipotle_df['quantity'],chipotle_df['item_price'],alpha=0.5)
```

```python
for item in chipotle_df['item_name'].unique():
    item_data = chipotle_df[chipotle_df['item_name'] == item]
    plt.scatter(item_data['quantity'], item_data['item_price'], alpha=0.5, label=item)

# Plot settings
plt.title('Scatter Plot of Quantity vs. Price with Item Name Legend')
plt.xticks(list(range(int(chipotle_df['quantity'].min()), int(chipotle_df['quantity'].max()) + 1)))
plt.xlabel('Quantity')
plt.ylabel('Price')
plt.legend()
plt.show()
```
